```python
#operator
x = 5
y = 5
z = 10

#and - returns True if both statements are true
if x > y and y > z:
    print("2 conditions are true")  #not print because condition is false

#or - returns True if one of the statements is true
if x > y or y > z:
    print("One of the conditions is true")  #not print because condition is false

#not - reverse the result, returns False if the result is true
if not x > y :
    print("Condition are not true")
```

```python
#nested if
j = 25
print("Nested if 1, j =", j)
if j > 10:
  print("Above ten")  #check if above 10
  if j > 20:    #execute if above 10, check if above 20
    print("and also above 20!")
  else:         #execute if above 10, execute if j is not above 20
    print("but not above 20.")
else:  #execute if j is not above 10
  print("Not above 10.")


j = 8
print("\nNested if 2, j =", j)
if j > 10:
  print("Above ten")  #check if above 10
  if j > 20:    #execute if above 10, check if above 20
    print("and also above 20!")
  else:         #execute if above 10, execute if j is not above 20
    print("but not above 20.")
else:  #execute if j is not above 10
  print("Not above 10.")
```

```python
#pass statement
#statement that execute nothing.
for i in range(2):
  pass  #do nothing, just loop through the range
  print(i)
print("Pass statement ended.")       
```

```python
#continue statement
#statement to skip loop
for i in range(3):
    if i == 1:
        continue  #skip the loop
    print(i)
print("Continue statement ended. 1 been skipped")
```

```python
#break statement
#statement to stop the loop
for i in range(4):
    if i == 2:
        break  #stop the loop
    print(i)
print("Break statement ended. Loop stopped at", i)
```

```python
for i in range(1, 10):
    if i <= 2:
        print(i, "- pass statement")
        pass #do nothing if i is 1/2, continue loop
    elif i == 4 or i == 5:
        print(i, "- continue statement")
        continue #skip loop if i is 4/5, proceed to i+=1
    elif i >= 8 and i <=9:
        print(i, "- break statement")
        break #stop loop if i is 8/9, end loop
    print("Loop", i, '\n---------')
print("Loop ended.")
```
