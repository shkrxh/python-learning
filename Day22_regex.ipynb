import re <br>
re.findall - get all match pattern <br>
re.match() - match only at the start <br>
re.search() - find match anywhere <br>
re.sub() - replace text <br>
re.split() - split by pattern <br>
re.compile - reuse pattern efficiently 


<b>Special Sequence</b><br>
\b - present in word <br>
\B - not present in word <br>
\d - contain digit (0-9) <br>
\D - not a digit <br>
\w - word character <br>
\W - not a word character <br>
\s - space, tab, newline <br>
\S - not a whitespace <br>
\b - word boundary <br>
\B - not a word boundary <br>
^  - start of string <br>
$  - end of string <br>
.  - any character except newline <br>

<br> <br>
<b>Sets</b><br>
[\*\*\*] - match any specific character/digit (insert character/digit) <br>
[*-*] - match lower case character/digit in between (inser character/digit) <br>
[a-zA-Z] - match lower case/upper case <br>
[*-*][*-*] - match two digit number 00-99 <br>
[^***] - except specific character (insert character) <br>

```python
#regular expression, RegEx
#pattern language for matching text
import re

#extract email in text
findall1 = "Contact me at test@email.com"
emails = re.findall(r"\S+@\S+", findall1)  #nonwhitespace + @ + nonwhitespace
print(emails)

#extract number in text
findall2 = "I have 2 apples and 10 bananas"
number = re.findall(r"\d+", findall2) #\d is digit, found more than 1 digit
print(number)
```

<b>Behind the code </b><br>
re.findall(r"\S+@\S+", text1)  
1. findall() - get all match pattern
2. start with r (raw string) to avoid conflict with special sequence
3. \S - return non whitespace
4. @  - contain @

re.findall(r"\d+", text2)
1. findall() - get all match pattern
2. start with r (raw string) to avoid conflict with special sequence
3. \d - contain digit
4. \+ \- one or more found

```python
#match at beginning
match1 = "Hello there! Nice to meet you."
hello = re.match(r"Hello", match1)
print(hello) #match
print(hello.group()) #display the match

match2 = "Good Morning Sunshine!."
hello = re.match(r"Hello", match2)
print(hello) #not match at beginning
```

```python
#searching and return the first match
search1 = "Bunga is 20 years old and Orkid is 18 years old"
number = re.search(r"\d+", search1)
print(number)
print(number.group()) #only return the frst match

search2 = "Bunga likes to eat apples"
apple = re.search(r"apple", search2)
print(apple is not None) #return True if match, return None if not match
```

```python
#replace
#re.sub(pattern, replacement, variable)
sub1 = "My number is 0112233 and 0445566"
number1 = re.sub(r"\d", "*", sub1) #replace each digit seperately
print(number1)
number2 = re.sub(r"\d+", "*", sub1) #replace entire number
print(number2)

sub2 = "Bunga love to eat at Coffee Bean"
name1 = re.sub(r"[eh]", "*", sub2) #replace e and h
print(name1)
name2 = re.sub(r"[e-h]", "*", sub2) #replace e - h
print(name2)

sub3 = "She      love   to   eat        apple."
space = re.sub(r"\s+", " ", sub3)
print(space)
```

```python
#split
split1 = "apple, mango, grape" 
fruit = re.split(r",", split1) #split the word by using , 
print(fruit)

split2 = "red orange purple"
colour = re.split(r"\s+", split2) #split the word using space
print(colour)
```

```python
#compile, reuse pattern
text1 = "I am 20 years old"
text2 = "I have 5 apples and 2 mangoes in my basket"

pattern1 = re.compile(r"\d+")  #find all digit
print(pattern1.findall(text1))
print(pattern1.findall(text2))

pattern2 = re.compile(r"\w+") # find all character
print(pattern2.findall(text1))
print(pattern2.findall(text2))
```
