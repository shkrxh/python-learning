```python
#match case (switch case)
#python 3.10+ only have match case statement 
i = 2
match i:
    case 1:
        print("Value is 1")
    case 2:
        print("Value is 2")
    case _: #default case, not match with any condition
        print("Value is other than 1 and 2")

status = "Success"
match status:
    case "Success":
        print("Operation is successful")
    case "Failed":
        print("Operation is failed")
    case _:
        print("Unknown status") 
```

```python
#operator match case
# | or

i = 3
match i:
    case 1 | 2:
        print("value is 1 or 2")
    case _:
        print("value is neither 1 nor 2")

status = "failed"
match status:
    case "Success" | "success":
        print("Operation is successful")
    case "Failed" | "failed":
        print("Operation is failed")
    case _:
        print("Unknown status") 
```

```python
#match case with if statement (as guard)

negara = "Korea"
bulan = 4
match bulan:
    case 3 | 4 | 5 if negara == "Korea":
        print("Musim bunga")
    case 6 | 7 | 8 if negara == "Korea":
        print("Musim panas")
    case _:
        print("Musin Luruh atau Sejuk")

p = "admin123"
match p:
    case p if len(p) < 8:
        print("Password terlalu pendek")
    case p if len(p) >= 8 and not any(c.isdigit() for c in p): #if no digit found
        print("Password harus mengandungi angka")
    case p if len(p) >= 8 and not any(c.isupper() for c in p): #if no uppercase found
        print("Password harus mengandungi huruf besar")
    case p if len(p) >= 8 and not any(c.isalnum() for c in p): #if no symbol found
        print("Password harus mengandungi simbol")
    case _:
        print("Password valid")
```
