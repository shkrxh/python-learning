```python
#tuple
#Tuple is a collection which is ordered and unchangeable. Allows duplicate members.

colour = ('merah', 'biru', 'putih', 'kuning', 'hijau', 'ungu')
print(colour)
print(type(colour))
print(len(colour))
```

```python
#accessing tuple -similar to list
print(colour)

#index start with 0
print(colour[1])

#access from the back
print(colour[-1])

#range
print(colour[1:3])  #include 1 and not include 3
print(colour[:3])  #display first 3 items
print(colour[3:])  #display item start from index 3 until the end
print(colour[-3:-1])  #include -3 and not -1 (from the back)
```

```python
#tuple cannot add/remove/change item - need to convert to list then convert back to tuple
#add item - convert to list
print('colour:', colour) #tuple
a = list(colour)  #convert to list
a += ['fuchsia', "magenta"]  #add item to the list
colour = tuple(a)  #convert back to tuple
print('update colour:', colour, '\n')

#add item - joining tuple
t1 = ('caramel', 'coffee')
print('t1:', t1)
t2 = ('mocha',)  #if 1 item need to put , so data type will be tuple
t1 += t2
print('update add t1:', t1)
```

```python
#remove item
print('t1:', t1)  #tuple
b = list(t1)  #convert to list
b.pop()  #remove item to the list
t1 = tuple(b)  #convert back to tuple
print('update remove t1:', t1)

del t1 #delete the tuple - error cannot print
```

```python
#change item in tuple - convert to list
tuple1 = ('item1', 'item2')  #tuple
print('tuple1:', tuple1)
list1 = list(tuple1)  #convert to list
list1[0] = 'additem'  #update item to the list
tuple1 = tuple(list1) #convert back to tuple
print('list1:', list1)
print('update tuple1:', tuple1)
```

```python
#unpacking tuple - assign
food = ('chips', 'piza', 'burger')

#unpacking a tuple - must have the same value of item 
print('unpack - same value')
(f1, f2, f3) = food
print('f1:', f1)
print('f2:', f2)
print('f3:', f3)

#unpack using asterisk - if variables do not have the same value of item
print('\nunpack  - do not have same value')
(f1, *f2) = food
print('f1:', f1)
print('f2:', f2)
```
