```python
#sets
#Set is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.

buah = {'epal', 'oren', 'anggur'}
print(buah)
print(type(buah))
print(len(buah))
```

```python
#accessing tuple - by loop
for x in buah:
    print(x)
print('\n')

#checking availability 
print("anggur" in buah)  #if anggur IN the set
print('anggur' not in buah)  #if anggur NOT IN the set
```

```python
#add item
#only for 1 item
buah.add('mangga')
print(buah)

#join set
buah2 = {'pisang','betik'} #can add any object (tuple, list, dictionary)
buah.update(buah2)
print(buah)
```

```python
#remove item
#remove specific item
buah.remove('oren')
print(buah)

#remove random item
x = buah.pop()  #item been remove
print('Item removed: ', x)
print('After pop: ', buah)

#clear the set
buah.clear()
print(buah)

del buah  #delete set - error cannot print
```
