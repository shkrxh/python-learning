```python
#list
#List is a collection which is ordered and changeable. Allows duplicate members.

transport = ['lori', 'bas', 'teksi', "kereta"]
print(transport)
```

```python
#acessing list
#index start with 0
print(transport[1])

#access from the back
print(transport[-1])

#range
print(transport[1:3]) #include 1 and not include 3
print(transport[:2]) #display 2 items from beginning
print(transport[2:]) #display item start from index 2 until the end
print(transport[-3:-1]) #include -3 and not include -1
```

```python
#add item at the end
#multiple item
transport.extend(["motor", "basikal"])  #can add any object (tuple, set, dictionary)
print(transport)

transport += ["kapal"]
print(transport)

#only for 1 item
transport.append('roket')
print(transport)

#add at specific index
transport.insert(6, 'bot')
```

```python
#remove item
transport.pop()
print(transport)

#remove specific index
transport.pop(1)
print(transport)

del transport[0]
print(transport)

#remove specific item
transport.remove('bas')
print(transport)

#empty the list
transport.clear()
print(transport)
```
