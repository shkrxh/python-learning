```python
#operators
a = 100
b = 200

print(f'a+b={a+b}')
print(f'a-b={a-b}')
print(f'a*b={a*b}')
print(f'a/b={a/b}')
print(f'a%b={a%b}') #modulus -
print(f'a**b={a**b}') #exponention
print(f'a//b={a//b}') #floor division - divide, round down and return integer
```

```python
#compare

print(a == b)  #equal to
print(a != b)  #not equal
```

```python
#membership operators

warna = ['putih', 'merah', 'kuning']
print ('merah' in warna)
print ('merah' not in warna)
```
