```python
#function with arguments and parameter
#number of parameter and argument must be the same
print("Example 1 (1 argument)")
def hi(name): #parameter
    print("Hi" , name)

hi("Orkid")  #argument

print("\nExample 2  (2 arguments)")
def add(a, b):  #parameters
    return a + b

calc = add(5, 3) #arguments
print(calc)

#default parameter and positional argument
#order of arguments must in correct order
print("\nExample 3 (default parameter and positional argument)")
def greet(name, greeting = "Hello"):
    print(greeting + ", " + name + "!")

greet("Orkid")  #uses default greeting
greet("Orkid", "Hi")  #overrides default greeting
greet("Bonjour", "Orkid")  #wrong position will cause different result

#keyword arguments
#order of arguments does not matter
print("\nExample 4 (keyword arguments)")
def animal(name, colour, type):
    print("I have a", colour, type, "named", name)

animal(name = "Oren", colour = "orange", type = "cat") #same position as parameter
animal(colour = "white", type = "dog", name = "Cookies") #different position
animal("Rainbow", colour = "multicolour", type = "fish") #positional argument must be before keyword argument
animal("Lucy", type = "rabbit",  colour = "black",) #positional argument must be before keyword argument
```

```python
#passing different data type as argument
print("Example 1")
def function(data):
    print("The data is", data)

function("Bunga") #string
function(123) #integer
function(["list1", "list2", "list3"]) #list
function((1, 2, 3)) #tuple
function({"set 1", "set 2", "set 3"}) #set
function({"name": "Orkid", "age": 1}) #dictionary

print("\nExample 2")
def function2(data2):
    for item in data2:
        print("The data is", item)

function2("Bunga") #string
function2(["list1", "list2", "list3"]) #list
```

```python
#return value
#can be store in variable or print directly
print("Example 1 (number)")
def add(a, b):  
    return a + b

calc = add(5, 3)
print(calc)

print("\nExample 2 (string)")
def greet(name):
    return "Hello, " + name + "!"

greeting = greet("Orkid")
print(greeting)

print("\nExample 3 (boolean)") 
def bool(num):
    return num > 0
print(bool(5))

print("\nExample 4 (list)")
def list(item1, item2, item3):
    return [item1, item2, item3]
my_list = list("apple", "banana", "cherry")
print(my_list)

#function with multiple return values
print("\nExample 5 (multiple return values as list)")
def return1():
    return ["Hello", 123] #return as list
output1 = return1()
print(type(output1))
print("Output1:", output1)
print("Output1[0]:", output1[0]) 
print("Output1[1]:", output1[1]) 

print("\nExample 6 (multiple return values as tuple)")
def calculate(a, b):
    sum = a + b
    difference = a - b
    product = a * b
    return sum, difference, product #return as tuple
sum_, difference, product = calculate(10, 5)  #variable and return value must be in the same order
print(type((sum_, difference, product))) 
print("Sum:", sum_)
print("Difference:", difference)
print("Product:", product)

#function with no return value
print("\nExample 7 (no return value)")
def message(message):
    print("The message is:", message) #the function does not return anything, only print
output = message("Hello, World!")
print("Output:", output)  
```

```python
#positional argument (position/order)
#keyword argument (name/label of parameter)

#positional only argument (/)
#arguments BEFORE / are positional only, cannot use keyword argument
print("Example 1 (positional only argument)")
def name(name, /):
    print("The name is", name)
name("Orkid")  #positional argument
#name(name = "Orkid")  #keyword argument will cause error

#keyword only argument (*)
#arguments AFTER * are keyword only, cannot use positional argument
print("\nExample 2 (keyword only argument)")
def name(*, name):
    print("The name is", name)
name(name = "Orkid")  #keyword argument
#name("Orkid")  #positional argument will cause error

#combine positional only and keyword only argument
print("\nExample 3 (combine positional only and keyword only argument)")
def name(name, /, *, greeting):
    print(greeting + ", " + name + "!")
name("Orkid", greeting = "Hello")  #positional argument and keyword argument
```
