```python
#change item
trp = ['lori', 'bas', 'teksi', 'kereta', 'motor', 'basikal', 'bot', 'kapal', 'roket']
print(trp)

trp[1] = "change01"
print(trp)

trp[2:4] = ("change02", 'change03')
print(trp)
```

```python
#sorting ascending
trp.sort()
print(trp)

#sorting descending
trp.reverse()
print(trp)
```

```python
#join lists
item1 = ['ayam', 'itik', 'lembu', 'kambing']
item2 = ['ikan', 'sotong']
haiwan = item1 + item2
print(haiwan)

b1 = ['epal', 'oren']
b2 = ['anggur', 'mangga', 'pisang']
for x in b2:
    b1.append(x)
print(b1)

#copy list
buah = b1.copy()
print(buah)
```

```python
#loop in the list
for x in buah:
    print(x)
    
print('\n') #\n = new line
    
for i in range(len(buah)):
    print(i, ':', buah[i])
    
print('b1:\t', b1)  #\t tab
print('buah:\t', buah)
```
