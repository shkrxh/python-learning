```python
#while statement
print("While loop")
i = 1
while i <= 3: #execute when condition is match
    print (i)
    i += 1
```

```python
#while loop with else statement
i = 1
while i <= 3:
    print (i)
    i += 1
else: #execute when condition is not match
    print("Loop is finished")
```

```python
#continue statement
#statement to skip loop
i = 1
while i <= 4:
    if i == 2:
        print("Skip 2")
        i += 1
        continue #skip the loop when condition is match
    print (i)
    i += 1
```

```python
#break statement
#statement to stop the loop
i = 1 
while i <= 5:
    print (i)
    if i == 3:
        print("Stop at", i)
        break #stop the loop when condition is match
    i += 1
```

```python
attempts = 0 

while attempts < 5:  
    attempts += 1
    password = input("Masukkan password: ") #Prompt user to enter password

    #if entered password is empty
    if password == "":
        print("Tak boleh kosong!")
        continue  #skip the loop and start next iteration

    #if entered password is "exit" or "Exit"
    if password == "exit" or password == 'Exit':
        print("User keluar")
        break  #exit the while loop

    #if password match
    if password == "Admin123!":
        print("Login berjaya")
        break  #exit the while loop
    
    #if password is not match
    print("Password salah") #execute when none of the conditions above are match

else:
    #Execute when while loop completes without encountering a break statement
    print("Terlalu banyak cubaan gagal")
```
