```python
#python lambda
#take any number of arguments, but can only have one expression
#quick function (lambda arguments: expression)
add = lambda a, b: a + b
print(add(2, 3))

#similar to
def add(a, b):
    return a + b
addition = add(2, 3)
print(addition)
```

```python
#map
#map(lambda arguments: expression, variable)
number = [3, 6, 9]
mapping = map(lambda x: x * 2, number) #map will apply the function in each of the number in the list
print(list(mapping)) #change the map to list - display result after mapping 

#filter
#filter(lambda arguments: expression, variable)
number = [2, 4, 5]
filtering = filter(lambda x: x % 2 == 0, number) #filter number that only return True
print(list(filtering)) #change the result into list

#sorted, dont forget the key
#sorted(variable, key = lambda arguments: expression)
#sort tuple 
print("\nExample 1 (sorting tuple):")
number = [(2, 8), (5, 1), (1, 5)]  #tuple
sorting = sorted(number, key = lambda x: x[0]) #sorting based on first item
#sorting = sorted(number, key = lambda x: x[1]) #sorting based on second item
print("Ascending sorting item x[0]:", sorting)
rev_sorting = sorted(number, key = lambda x: x[0], reverse = True)
print("Descending sorting item x[0]:", rev_sorting)

print("\nExample 2 (sorting by length of string):")
colour = ["Magenta", "White", "Blue"]
sorting2 = sorted(colour, key = lambda x: len(x))
print(sorting2)
```

```python
#map
print("Exercise map 1:")
animal = ["dolphin", "snake", "ant"]
mapping1 = map(lambda x: x.upper(), animal)
print(list(mapping1))

print("Exercise map 2:")
name = ["bunga", "orkid"]
mapping2 = map(lambda x: "Hi " + x.title() , name)
print(list(mapping2))

#filter
print("\nExercise filter 1:")
country = ["Malaysia", "China", "Brunei"]
filter1 = filter(lambda x: len(x) > 5, country)
print(list(filter1))

print("Exercise filter 2:")
flower = ['Peony', 'Bluebell']
filter2 = filter(lambda x: x.startswith('B'), flower)  #find flower that start with B
print(list(filter2))

#sorted
print("\nExercise sorted 1:")
colour = ['red', 'purple', 'blue']
sorted1 = sorted(colour, key = lambda x: len(x)) #length of string
print(list(sorted1))
sorted2 = sorted(colour) #ascending order
print(list(sorted2))

print("Exercise 2")
name = [{"name": "Bunga", "age": 20},
        {"name": "Orkid", "age": 20},
        {"name": "Blossom", "age": 21}]
sorted3 = sorted(name, key = lambda x: x["name"])  #sorted by name
print(list(sorted3))
sorted4 = sorted(sorted3, key = lambda x: x["age"])  #sorted by age
print(list(sorted4))
sorted_result = sorted(name, key=lambda x: (x["age"], x["name"]))  #better way
print(sorted_result)
```
