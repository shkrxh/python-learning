<b>F-String </b><br>
f"....{variable}....."
1. f - put infront of the string
2. {variable} - placeholder or/and modifier to format the value

```python
#easy to combine different data type
name = "Bunga"
age = 20
combine1 = f"My name is {name}. I am {age} years old"
print(combine1)
```

```python
#can do calculation
age = 20
calc1 = f"Orked is 2 years older. She is {age + 2} years old"
print(calc1)

price = 2.50  #price per kg
weight = 1.3  #weight of fruit
fruit = "apple"
calc2 = f"Bunga bought {fruit} at the market. The price is RM{price * weight}"
print(calc2)
```

<b>Format number</b><br>
{variable:.2f} <br>
1. variable - variable <br>
2. : - start formatting <br>
3. .2 - 2 decimal place (can change) <br>
4. f - float (data type)

```python
#format number
weight = 0.1522
price = 100
fornum1 = f"Today's gold weight is {weight:.3f}g for RM{price:.2f}"
print(fornum1)

fornum2 = f"{weight:.0%}"
print("No decimal place percentage: ", fornum2) 
#.0 - no decimal place
#% - convert to percentage

fornum3 = f"{weight:.2%}"
print("2 decimal place percentage: ", fornum3) 
#.2 -2 decimal place
#% - convert to percentage

num = 9988776655
print(f'Number with thousand seperator: {num:,}')
```

```python
#if statement in fstring
#f" {"Result if true" condition "Result if false"} ""
score = 80
print(f"Result: {"Pass" if score >=70 else "Fail"}")

score = 80
attendance = 80
result = f"Pass" if score >= 70 and attendance > 80 else "Fail"
print("Result:", result)

score = 80
result = f"A" if score >= 90 else "B" if score >= 70 else "C" if score >=50 else "D"
print("Result :", result)

score = 80
result = f"Your score: {score}. {'Pass' if score >= 70 else 'Fail'}" #outer string "", inner string '' or vice versa
print(result)
```
