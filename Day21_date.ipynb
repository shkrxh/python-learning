import datetime <br>
date - year, month, day <br>
time - hours, minutes, seconds <br>
datetime - combine date and time <br>

```python
#python date
#import datetime module
import datetime

#get current datetime
now = datetime.datetime.now()
print("Datetime :", now)  #yy-mm-dd hh:mm:s.microsecond

#get today date
date = datetime.date.today()
print("Date :", date) #yy-mm-dd
```

```python
#create specific date
dt1 = datetime.date(2026, 4, 1) 
print(dt1)

#create specific datetime
dt2 = datetime.datetime(2026, 4, 1, 18, 25, 40) 
print(dt2)
```

<b>Date format code </b><br>
%Y - year (2026) <br>
%m - month (1-12)<br>
%b - month (Jan - Dec)<br>
%B - month (January - December)<br>
%a - weekday (Mon - Sun) <br>
%A - weekday (Monday - Sunday) <br>
%d - day <br>
%H - hour (24h) <br>
%I - hour (12h) <br>
%M - minute <br>
%S - second <br>


```python
#format datetime
now1 = datetime.datetime.now()
print("Now :", now1)  #yy-mm-dd hh:mm:s.

now2 = now.strftime("%d-%m-%Y") #dd-mm-yy
print("dd-mm-yy :", now2)

now3 = now.strftime("%A") #Weekday in full version
print("Day :", now3)

now4 = now.strftime("%B") #Month in full version
print("Month :", now4)
```

```python
now = datetime.datetime.now()
print("Today datetime:", now)

#add time
tomorrow1 = now + datetime.timedelta(days = 1)
print("Tomorrow datetime:", tomorrow1)

tomorrow2 = now + datetime.timedelta(days = 1)
print("Tomorrow date:", tomorrow2.date())

next_5days = now + datetime.timedelta(days = 5)
print("Next 5 days datetime:", next_5days)

#subtract time
yesterday = now - datetime.timedelta(days = 1)
print("Yesterday date format:", yesterday.strftime("%d-%B-%Y"))

past_5days = now - datetime.timedelta(days = 5)
print("Past 5 days datetime:", past_5days)
```

```python
start_date = datetime.datetime(2026, 3, 21)
end_date = datetime.datetime(2026, 4, 18)

date1 = datetime.datetime(2026, 3, 31)
print(date1.date())
if start_date <= date1 <= end_date:
    print("Selamat Hari Raya!")
elif date1 <= start_date:
    print("Not yet Raya")
else:
    print("Raya already finish")

date2 = datetime.datetime(2026, 4, 23)
print(date2.date())
if start_date <= date2 <= end_date:
    print("Selamat Hari Raya!")
elif date2 <= start_date:
    print("Not yet Raya")
else:
    print("Raya already finish")

date3 = datetime.datetime(2026, 2, 23)
print(date3.date())
if start_date <= date3 <= end_date:
    print("Selamat Hari Raya!")
elif date3 <= start_date:
    print("Not yet Raya")
else:
    print("Raya already finish")

different_day = end_date - start_date
print("end_date - start_date =", different_day.days)
```
