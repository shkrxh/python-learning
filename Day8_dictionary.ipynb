```python
#dictionary
#Dictionary is a collection which is ordered** and changeable. No duplicate members.
#key: value

data = {"nama": "Bunga",
        "umur": 20,
        'jantina': 'perempuan'}

print(data)
print(type(data))
print(len(data))
```

```python
#accessing dictionary 
print(data['nama'])

a = data.get('umur')
print(a)    

#print key
print('\nPrint key')
for x in data:
    print(x)

#print value
print('\nPrint value 1')
for x in data:
    print(data[x]) 

print('\nPrint value 2')
for x in data.values():
    print(x)   

#print key and value
print('\nPrint key and value')
for x, y in data.items():
  print(x, ':', y)
```

```python
#add item
data['aliran'] = 'sains'
print(data)

data.update({'hobi': 'melukis'})
print(data)
```

```python
#change item in dictionary
data['umur'] = 18
print(data)

data.update({'aliran': 'komputer'})
print(data)
```

```python
#remove item
#remove specified item
data.pop('umur')
print(data)

#remove last item
data.popitem()
print(data)

#clear dictionary
data.clear()
print(data)

del data  #delete dictionary - error cannot print
```

```python
#copy dictionary
a = {'buah': 'mangga', 
     'warna': 'kuning'}
copy1 = dict(a)
print(copy1)
```

```python
#nested dictionary
#declare 1 dictionary with 2 dictionaroies inside
buah = {
    'buah1': {
        'nama': 'mangga',
        'warna': 'kuning'
    },
    'buah2': {
        'nama': 'epal',
        'warna': 'merah'
    }
}
print(buah)
print('buah1 nama:', buah['buah1']['nama'])

#print key (buah1, buah2) and value (its dictionary)
print('\nPrint key and value')
for x, y in buah.items():
    print(x, ':', y)

#loop through each item in dictionaries
(print('\nPrint key and value in each dictionary'))
for x, obj in buah.items(): #x is key (buah1, buah2), obj is value (its dictionary)
    print(x, ':')
    for key, value in obj.items(): #loop in the dictionary
        print(key, ':', value)
```
