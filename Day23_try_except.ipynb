try: <br>
&nbsp;&nbsp;&nbsp;&nbsp;code that might cause error <br>
except: <br>
&nbsp;&nbsp;&nbsp;&nbsp;code to run when error occur <br>
else: <br>
&nbsp;&nbsp;&nbsp;&nbsp;code run when no error occur <br>
finally:<br>
&nbsp;&nbsp;&nbsp;&nbsp;code that always run <br>

```python
#try except
#used to handle errors (exceptions) to avoid program stop
a = 10
b = 0
print("Cause error")
print("10 / 0 = error because divided by 0")

print("\nWay to handle error")
try:
    print(a/b)
except:
    print("Error!")
```

```python
#build in exception name (ValueError, ZeroDivisionError, IndexError)
try:
    num = int(input("Enter a number: "))
    output = 10/ num
    print("Output", output) #cause error
except ValueError:  #build in exception name
    print("Error input. Input must be number!")
except ZeroDivisionError: #build in exception name
    print("Cannot divide by zero!")
```

```python
try:
    numbers = [10, 20, 30]
    index = int(input("Enter index: "))
    print(numbers[index])  #cause error
except IndexError:  #build in exception name
    print("Index out of range!")
    print("Index only have", len(numbers))
except ValueError:  #build in exception name
    print("Please enter a valid number!")
else: #execute when no error occur
    print("Index :", index, "\tValue :", numbers[index])
finally:  #execute either error or not
    print("Index entered :", index)
```

```python
#raise an exception
#manually raise exception
try:
    age = int(input("Enter your age: "))
    if age < 0 :
        raise ValueError("Age cannot be negative.")  #manually raise error based on condition
    print("Your age is", age)
except ValueError:
        print("Invalid input. Please enter numbers only")
```

```python
def operator(op, valA, valB):
    if op == 1: #add
        output = valA + valB
        print(valA, "+", valB, "=", output)
    elif op == 2:  #sub
        output = valA - valB
        print(valA, "-", valB, "=", output)
    elif op == 3:  #multiply
        output = valA * valB
        print(valA, "*", valB, "=", output)
    elif op == 4:  #divide
        try:
            output = valA / valB
            print(valA, "/", valB, "=", output)
        except ZeroDivisionError: #build in exception name
            print("Cannot divide by zero!")
    else:
        print("Wrong input for operation. Only 1 - 4")

while True:  #looping until user choose exit
    print("\nSelect an operation. Enter a number between 1 to 4.\n" 
        "1 - Addition \n"
        "2 - Subtraction \n" 
        "3 - Multiplication \n" 
        "4 - Division \n"
        "0 - Exit \n")

    try:
        operation = int(input("Enter your operation code: "))
        print("You choose option", operation)

        if operation == 0:
            print("Goodbye")
            break  #will exit the loop when user entered 0 to exit

        valA = int(input("Enter your first value: "))
        print("First Value", valA)

        valB = int(input("Enter your second value: "))
        print("Second Value", valB)

        operator(operation, valA, valB)

    except ValueError:
        print("Invalid input. Please enter numbers only")
```
