```python
#range
#range(start, stop, step)

#one argument
print("Example 1 (One argument):")
for i in range(3):  #start 0 by default
    print(i)

print("\nExample 2 (Two arguments):")
for i in range(1, 4):  # i >= 1 and i < 4
    print(i)

print("\nExample 3 (Three arguments):")
for i in range(1, 5, 2): # i >= 1 and i < 5, increase by 2
    print(i)
```

```python
number = range(10) # 0-9
print(number)
#display range
print(list(number))
#display length of range
print("Length of number:", len(number))

print("\n")
#slicing range
print("Index 3:", number[3])
print("Index 3 from behind:", number[-3])
print("Index 1 until 3:", list(number[1:4])) #include 1 and not include 4
print("First 3 index:", list(number[:3])) #display 3 item from beginning
print("Index 3 until the end:", list(number[3:])) #display item start from index 3 until the end
print("Last index 4 until last index 2 :", list(number[-4:-1])) #include -4 and not include -1
print("Not include 3 last index:", list(number[:-3])) #display 3 item from beginning
print("Last index 3 until the end:", list(number[-3:])) #display item start from index 3 until the end

```

```python
#membership testing
number = range(10) # 0-9
print(8 in number)
print(12 in number)
```
