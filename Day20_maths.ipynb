```python
#build-in maths
#min() - smallest value
#max() - biggest value
#sum() - total
number = [3, 7, 10, 5, 1]
print("Min: ", min(number))
print("Max: ", max(number))
print("Sum: ", sum(number))
```

```python
#abs() - absolute value, positive number
print(abs(-2))

number = [-10, -4, -9]
for i in number:
    print("Before:", i, "\tAfter:", abs(i))
```

```python
#round() - rounding number
number = 1234.5678
print("Nearest integer:", round(number))
print("Nearest 1 decimal place:", round(number, 1))
print("Nearest 2 decimal place:", round(number, 2))
print("Rounding number tenth:", round(number, -1))
print("Rounding number hundredth:", round(number, -2))
```

```python
#pow() - power
#2 arguments (a ^ b)
power1 = pow(2, 3)  #2 * 2 * 2
print("2^3 =", power1)

power2 = pow(1, 0)
print("1^0 =", power2)

#3 arguments (a ^ b % c)
power3 = pow(2, 3, 3)  #(2 * 2 * 2) % 3
print("2^3%3 =", power3)

power4 = pow(2, 5, 4)  #(2 * 2 * 2* 2 * 2) % 4
print("2^5%4 =", power4)
```

```python
#divmod(a, b) - quotient, remainder
# a//b  a%b
divmod1 = divmod(10, 3)  #10 / 3
print(divmod1)  #3 remainder 1

divmod2= divmod(20, 4)  #20 / 2
print(divmod2)  #5 remainder 0
d, r= divmod(20, 4)  #20 / 2
print("Divide:", d)
print("Remainder:", r)
```

```python
import math  
#math.sqrt() - square root
sqrt1 = math.sqrt(25)
print("Square root 25 =:", sqrt1)

sqrt2 = math.sqrt(5)
print("Square root 5 =:",sqrt2)
```

```python
#math.floor() - round down
floor1 = math.floor(1.22)
print("Round down 1.22 =", floor1)

floor2 = math.floor(5.66)
print("Round down 5.66 =", floor2)

floor3 = math.floor(-2.33)
print("Round down -2.33 =", floor3)

floor4 = math.floor(-6.77)
print("Round down -6.77 =", floor4)

print("\n")
#math.ceil() - round up
ceil1 = math.ceil(1.22)
print("Round up 1.22 =", ceil1)

ceil2 = math.ceil(5.66)
print("Round up 5.66 =", ceil2)

ceil3 = math.ceil(-2.33)
print("Round up -2.33 =", ceil3)

ceil4 = math.ceil(-6.77)
print("Round up -6.77 =", ceil4)
```

```python
#trigonometry
#python use radians

#degree to radian
degree1 = 90
rad = math.radians(degree1) 
print("rad 90 =", rad)

#radianto degree
radians1 = 1
deg = math.degrees(radians1)
print("Degree 100 =", deg)

```

```python
#math.sin()
sin1 = math.sin(90)
print("Sin 90 =", sin1)


#math.cos(x)
cos1 = math.cos(90)
print("Cos 90 =", cos1)


#math.tan(x)
tan1 = math.tan(90)
print("Tan 90 =", tan1)
```

```python
#constant
#math.pi - π
print(math.pi)

#math.e - Euler’s number
print(math.e)
```
