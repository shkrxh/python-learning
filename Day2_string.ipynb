```python
#start with 0
x = 'Nama saya Bunga.'

print('len(x):', len(x))
print('x[0]:', x[0])  #N
print('x[5]:', x[5])  #s
```

```python
#slicing
print('x[1:3]:', x[1:3])  #not include 3
print('x[:3]:', x[:3]) #not include 3
print('x[3:]:', x[3:])  #start from 3 to the end
print('x[-6:-1]:', x[-6:-1]) #start from back and not inclde 1
```

```python
#modify string
print('x.upper():', x.upper())  #uppercase
print('x.lower():', x.lower())  #lowercase
print('x.title():', x.title())  #title case
print('x.replace("B", "z"):', x.replace("B", "z"))  #replace, case sensitive
print('x.split(" "):', x.split(" "))  #result will be in list
```

```python
#fstring - format string
#boleh nak sumkan, format tarikh (%B-bulan, %d-date, %Y-year)
#tak boleh direct guna \

item = 'ayam'
harga = 5.55
print(f"Harga {item}: {harga}/kg")

itik = 3
print(f'Harga itik: {itik:.2f}')  #kena ada . selepas : so jadi 2dp

lembu = 1234567
print(f'Ada coma seperator: {lembu:,}')
```

```python
#concatenate string
a = "Selamat"
b = "Malam"
c = a + " " + b
print(c)
```
