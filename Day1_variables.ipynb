```python
#variables
x = 5  #int
y = "Sunshine"  #string
z = 'Moonpie'  #can use '...' or "..."

print(x)
print(y)
print(z)
```

```python
#get data type
print(type(x))
print(type(y))
```

```python
x = str(3)
y = int(3)
z = float(3)

print (x)
print (y)
print (z)

#get data type
print(type(x))
print(type(y))
print(type(z))
```

```python
#multiple variables
x, y, z = "Ayam", "Itik", "Angsa"

print (x)
print (y)
print (z)
```

```python
x = y = z = 'Udang'

print (x)
print (y)
print (z)
```

```python
x, y, z = 10, 25, 3
print ("x:", x)
print ("10+25=", x + y)
print ("25-3=", y - z)
```

```python
#variables in tuple
haiwan = ["lembu", "kambing", "kerbau"]
x, y, z = haiwan

print (x)
print (y)
print (z)
print ('x + y: ' + x + y) #using + have no space
print ('x , y: ' + x , y) #using , have space
```
