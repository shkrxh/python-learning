```python
#if statement
a = 20
b = 5

#condition
print(a == b)  #equal
print(a != b)  #not equal
print(a < b)   #less than
print(a <= b)  #less than or equal to
print(a > b)   #greater than
print(a >= b)  #greater than or equal to 
```

```python
#indentation and : is important
print('Contoh 1')
if a > b:   
    print('a lebih besar dari b') 
elif a == b:  #else if
    print('a sama dengan b') 
else:
    print('a lebih kecil dari b')   

print('\nContoh 2')
if a > b:   
    print('a lebih besar dari b') 
    print("Nilai a: ", a)
else:
    print('a lebih kecil dari b')
    print("Nilai b: ", b) 
```

```python
#short hand if
if a > b: print('a lebih besar dari b')

#short hand if else
print('a lebih besar dari b') if a > b else print('a lebih kecil dari b')

besar = a if a > b else b
print("Angka besar", besar)
```
