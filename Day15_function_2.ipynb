```python
#python *args (arbitrary number of arguments)
#allow function to accept any number of arguments
#*args is a tuple of arguments passed to the function
print("Example 1:")
def args_function(*name): 
    print(type(name))  #tuple
    print(name)
    print(name[0])  #accessing first argument
    print(name[1])  #accessing second argument
args_function("Bunga", "Orkid")

print("\nExample 2:")
def args_function2(greeting, *name):
    for name in name:
        print(greeting, name) #greeting is a normal argument, *name is *args
args_function2("Hello","Bunga", "Orkid")

print("\nExample 3:")
def args_function3(*marks):
    total = 0
    for mark in marks:
        total += mark
    return total 
print("Total marks:", args_function3(85, 90, 78))
```

```python
#python **kwargs (keyword arguments)
#allow function to accept any number of keyword arguments
#**kwargs is a dictionary of keyword arguments passed to the function
print("Example 1:")
def kwargs_function(**kwargs):
    print(type(kwargs))  #dictionary
    print(kwargs)
    print(kwargs["name"])  #accessing value of 'name' key
    print(kwargs["age"])   #accessing value of 'age' key
kwargs_function(name = "Bunga", age = 20)

print("\nExample 2:")
def kwargs_function2(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value) #greeting is a normal argument, **kwargs is **kwargs
kwargs_function2(name = "Bunga", age = 20)

print("\nExample 3:")
def kwargs_function3(**kwargs):
    total = 0
    for mark in kwargs.values():
        total += mark
    return total 
print("Total marks:", kwargs_function3(a = 85, b = 90, c = 78))
```

```python
#combine *args and **kwargs
#must in order: regular arguments, *args, **kwargs
print("Example 1:")
def combined_function(greeting, *args, **kwargs):
    print(greeting)
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)
combined_function("Hello", "Bunga", "Orkid", name = "Bunga", age = 20)

print("\nExample 2:")
def combined_function2(*args, **kwargs):
    total = sum(args)

    if kwargs.get("discount"):
        total *= 0.9  #if True, apply discount
        print("Discount applied!")
    return total
print("Total:", combined_function2(85, 90, discount = True))
print("Total:", combined_function2(85, 90, discount = False))
```

```python
#unpacking arguments
print("Example 1 (*args):")
def unpacking_argument(greeting, name):
    print(greeting, name)
args = ("Hello", "Bunga")
unpacking_argument(*args) #unpacking tuple into positional arguments

print("\nExample 2 (**kwargs):")
def unpacking_arguments2(name, age):
    print("Name:", name)
    print("Age:", age)
unpack_dict = {"name": "Bunga", "age": 20} #assign to a variable
unpacking_arguments2(**unpack_dict) #unpacking dictionary into keyword arguments
print(unpack_dict)

#packing arguments
print("\nExample 1 (*args):")
def packing_arguments(*args):
    for arg in args:
        print(arg)
packing_arguments("Bunga", "Orkid") #packing arguments into tuple

print("\nExample 2 (**kwargs):")
def packing_arguments2(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)
packing_arguments2(name = "Bunga", age = 20) #packing arguments into dictionary
```

```python
#*args
def find_max(*args):
    max_num = args[0] #assign first argument as max_num
    for num in args:
        if num > max_num: #compare each number with max_num
            max_num = num #update max_num if current number is bigger
    return max_num
maximum = find_max(6, 7, 10, 15, 3)
print("Biggest number in list:", maximum)

#**kwargs
def total_price(**kwargs):
    total = 0
    for price in kwargs.values():  #only use values
        total += price
    return total
print("Total price:", total_price(buku = 20, pen = 5, bag = 50))

#combining *args and **kwargs
def counting(*args, **kwargs):
    output = 0
    if kwargs.get("math") == "add": #if math is add, sum all numbers
        output += sum(args)
    elif kwargs.get("math") == "multiply": #if math is multiply, multiply all numbers
        output = 1
        for num in args: #if math is not add, multiply all numbers
            output *= num
    return output
print("Add:", counting(1, 6, 5, math = "add"))
print("Multiply:", counting(1, 6, 5, math = "multiply"))
```
