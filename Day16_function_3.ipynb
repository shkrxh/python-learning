<b>LEGB Rule</b> <br>
Local       : Inside the current function <br>
Enclosing   : Inside enclosing functions (from inner to outer)  <br>
Global      : At the top level of module  <br>
Built       : In python's built -in namespace  

```python
#local scope
#variable that is available only inside the function
def local_scope():
    x = 10
    print("Inside the function, x =", x)
local_scope()
#print(x) #error

#function inside a function, still has access to the local variable of the outer function
def outer_local_scope1():
    x = 20
    def inner_function():
        print("Inside the inner function, x =", x) #still has access to x
    inner_function()
outer_local_scope1()
```

```python
#global scope
#variable that is available throughout the program
x = 30
def global_scope():
    print("Inside the function, x =", x) #accessing global variable
global_scope()
print("Outside the function, x =", x) #accessing global variable
```

```python
#global scope and local scope together
x = 40
def global_local_scope():
    x = 50  #local variable that shadows the global variable
    print("Inside the function, x =", x) #accessing local variable
global_local_scope()
print("Outside the function, x =", x) #accessing global variable
```

```python
#global keyword
#change global variable in function
x = 100
def global_keyword():
    global x  #change global variable 
    x = 10  #modify global variable
    print("Inside the function, x =", x)

print("Initial value, x =", x)
global_keyword()
print("Final value, x =", x)
```

```python
#nonlocal keyword
#used inside nested function
def outer():  #enclosing for all function in outer()
    x = 1
    print("\n[1] OUTER start → x =", x)

    def middle():
        print("\n[3] MIDDLE start → x =", x)

        def inner():
            nonlocal x  #want to use from outside of the scope
            print("\n[5] INNER start → x =", x)

            x = 99  #modify value
            print("[6] INNER after change → x =", x)

        print("[4] MIDDLE before calling inner() → x =", x)
        inner()
        print("[7] MIDDLE after inner() → x =", x)

    print("[2] OUTER before calling middle() → x =", x)
    middle()
    print("\n[8] OUTER after middle() → x =", x)

outer()
```

```python
x = 100  #global scope

def outer():
    x = 10  #enclosing
    print("\n[1] OUTER start → x =", x)

    def middle():
        y = 5  #local in middle()
        print("\n[3] MIDDLE start → x =", x, ", y =", y)

        def inner():
            nonlocal x   #want to use from outer()
            global z     #create and change global variable

            z = 999      #global keyword
            x = 50       #nonlocal keyword (change value x that in outer())

            w = 1        #local scope in inner()s
            print("\n[5] INNER → x =", x, ", y =", y, ", w =", w, ", z =", z)

        print("[4] MIDDLE before inner() → x =", x, ", y =", y)
        inner()
        print("[6] MIDDLE after inner() → x =", x, ", y =", y)

    print("[2] OUTER before middle() → x =", x)
    middle()
    print("[7] OUTER after middle() → x =", x)

outer()

print("\n[8] GLOBAL x =", x)  #global scope
print("[9] GLOBAL z =", z)   #global keyword that create in inner()
```
