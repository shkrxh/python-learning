```python
#for loop
print('For loop in string')
a = "Hello"
for i in a:
    print(i) 

print('\nFor loop in list')
b = ['list1', 'list2']
for i in b:
    print(i)

print('\nFor loop in tuple')
c = ('tuple1', 'tuple2')
for i in c:
    print(i)

print('\nFor loop in set')
d = {'set1', 'set2'} 
for i in d:
    print(i)

print('\nFor loop in dictionary')  
e = {'dict1': 'value1', 'dict2': 'value2'}
for i in e:
    print(i)
```

```python
#for loop in range()
print('For loop in range(3)')
for i in range (3):  #i < 3, start with 0 by default, increase by 1 by default
    print(i)

print('\nFor loop in range(1, 3)')
for i in range (1, 3):  #i >= 1 and i < 3
    print(i)

print('\nFor loop in range(1, 10, 2)')
for i in range (1, 9, 2):  #i >= 1 and i < 9, increase by 2
    print(i)
```

```python
#for loop with else statement
for i in range(3):
    print(i)
else: #execute when loop is over
    print('Loop is over')
```

```python
#continue statement
#statement to skip loop
print('Example 1')
for i in range(4):
    if i == 2:
        print("Skip 2")
        continue #skip the rest of the code in the loop when i == 2
    print(i)

print('\nExample 2')
negara = ["Korea", "Jepun", "China"]
for i in negara:
    if i == "Jepun":
        print("Negara Jepun")
        continue
    print(i)
    
```

```python
#break statement
#statement to stop the loop
print('Example 1')
for i in range(4):
    if i == 2:
        print("Stop at 2")
        break #stop the loop when i == 2
    print(i)

print('\nExample 2')
negara = ["Korea", "Jepun", "China"]
for i in negara:
    if i == "Jepun":
        print("Stop at Jepun")
        break
    print(i)

#for loop with else statement and break statement
#else statement will not execute when break statement is executed
print('\nExample 3')
for i in range(4):
    if i == 2:
        print("Stop at 2")
        break #stop the loop when i == 2
    print(i)
else:
    print("Loop completed without break")
```

```python
#pass statement
#statement that execute nothing.
#empty for loop cause error, use pass to avoid error
for i in range(2):
  pass  #do nothing, just loop through the range
  print(i)
print("Pass statement ended.")    
```

```python
#nested loop
print('Example 1')
for i in range(3):  #outer loop
    for j in range(2):  #inner loop
        print(f'i: {i}, j: {j}')  

print('\nExample 2')
baju = ['baju1', 'baju2']
size = ['S', 'M', 'L']
for i in baju:
    print(i)
    for j in size:
        print(i, j)
```
