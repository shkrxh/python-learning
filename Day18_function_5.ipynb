```python
#python recursion
#function call itself to solve problem
#base case - condition to stop (avoid infinite loop)
#recursive case - call itself for smaller input 

#factorial, n! = n x (n - 1)!
def factorial(n):
    if n == 1:  #base case
        return 1
    return n * (factorial(n - 1))  #recursive case
print(factorial(4))
```

<b>Behind the code (call stack)</b><br>
= 4 * factorial(3)<br>
= 4 * 3 * factorial(2)<br>
= 4 * 3 * 2 * factorial(1)<br>
= 4 * 3 * 2 * 1<br>
= 24

```python
print("Find maximum number in list")
def find_max(numbers):
  if len(numbers) == 1:  #if only 1 number in list, hence it is the max
    return numbers[0]
  else:
    max_of_rest = find_max(numbers[1:]) #[7,2,9,1], [2, 9, 1], [9, 1], [1]
    return numbers[0] if numbers[0] > max_of_rest else max_of_rest

num = [3, 7, 2, 9, 1]
print(find_max(num))
```

```python
#python generators (yield)
#return value one by one
#pause at each yield, continue when next()
def count():
    for i in range(3):
        yield i

g = count()

print(next(g)) #0
print(next(g)) #1, continue from previous
print(next(g)) #2, continue from previous
```

```python
#fibonacci, f(n) = f(n-1) + f(n-2)
def fibonacci():
   a, b = 0, 1
   while True:
    yield a  #(0)
    a, b = b, a + b  #(1, 0 + 1)

#Get first 10 fibonacci numbers
gen = fibonacci()
for _ in range(7):
   print("Yield", _, ": ", next(gen))
```

<b>Behind the code</b><br>
a, b = 0, 1     a = 0, b = 1    output: 0 <br><br>

a, b = b, a + b<br>
(1, 0 + 1)      a = 1, b = 1    output: 1<br>
(1, 1 + 1)      a = 1, b = 2    output: 1<br>
(2, 1 + 2)      a = 2, b = 3    output: 2<br>
(3, 2 + 3)      a = 3, b = 5    output: 3<br>
(5, 3 + 5)      a = 5, b = 8    output: 5<br>
(8, 5 + 8)      a = 8, b = 13    output: 8<br>
